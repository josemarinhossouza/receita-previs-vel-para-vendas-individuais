```python
# =============================================================
# RECEITA PREVISÍVEL - Baseado no livro de Aaron Ross
# =============================================================
# Este código é um protótipo que projeta a receita esperada 
# aprentada por Aaron Ross em seu livro Receita Previśivel

# COMO FUNCIONA:
#   - São vendas B2B, o processo da venda é construído por funil de vendas até o cliente ser categorizado como comprador que irá fechar a venda,
#   - O indicador está adaptado para realidade um único vendedor,
#   - Pode adaptar para uma equipe de vendas,
#   - O vendedor trabalha leads em um processo consultivo contínuo, onde o contato é cultivado até que o lead demonstre real  
#       intenção de fechar negócio. Somente nesse momento ele é contado como lead qualificado ativo (LQ). Isso elimina leads 
#       frios do cálculo e torna a projeção mais confiável,
#   - Os VALORES FIXOS vêm do histórico de vendas, devem ser alterados quando o histórico apresentar mudança significativa,
#   - O ÚNICO valor que você muda no dia a dia é o LQ (leads ativos agora)
#   - O código apresenta duas receitas previsíveis: 
#       - R cilo, é a receita previsível no intervalo de uma venda para outra (dias, semanas, meses), venda de qualquer cliente.
#       - R 30 dias: é a receita previsível ao fim do mês.

# FÓRMULAS USADAS:
#   E  (Eficiência)         = Vendas realizadas / Leads históricos
#   TM (Ticket Médio)       = Total em R$ / Quantidade de vendas
#   PP (Período Projetado)  = 30 dias / Ciclo médio
#   R ciclo                = LQ × E × TM
#   R 30 dias               = LQ × E × TM × PP
# =============================================================
```

```python
import pandas as pd
import math
```

```python
# Total de leads que passaram pelo funil no período histórico ao fim do ciclo (gap entre uma venda e outra)
# (esse número vem do seu CRM ou controle de leads)

leads_historico = 20
 
# Período base em dias (normalmente 30 dias = 1 mês)
periodo_base = 30
```

```python
# O CSV com registro das vendas precisa ter três colunas:
#   data      → data do dia
#   venda     → valor da venda naquele dia (deixe vazio se não houve venda)
#   intervalo → dias desde a última venda
 
dados = pd.read_csv('/home/jose-marinho/Documentos/Python/Projetos/GitHub/Receita Previsível/dados_vendas.csv')
 
print(f"  Total de linhas no arquivo: {len(dados)}")
```

```python
# =============================================================
# CALCULAR OS VALORES DA RECEITA PREVISÍVEL, SÃO FIXOS A PARTIR DO HISTÓRICO
# (esses valores só mudam quando você atualiza o histórico)
# =============================================================
 
# --- Quantidade de vendas (V) ---
# Conta quantas linhas têm um valor de venda preenchido
V = dados["venda"].notna().sum()
 

# Soma todos os valores de venda
total_reais = dados["venda"].sum()
 

# Pega só os intervalos preenchidos (dias entre uma venda e outra)
intervalos = dados["intervalo"].dropna() # intervalos
soma_ciclos = intervalos.sum()       # soma dos intervalos
qtd_ciclos  = intervalos.count()    # quantidades de ciclos
ciclo_medio = soma_ciclos / qtd_ciclos  # média dos ciclos
 
# E: Eficiência (taxa de conversão)
# Quantas vendas foram feitas por lead que entrou no funil
e = V / leads_historico  # ex: 7 / 20 = 0,35 → 35%
 
# TM: Ticket Médio
# Valor médio por venda
tm = total_reais / V  # ex: 10.000 / 7 = 1.428,57
 
# PP: Período Projetado
# Quantos ciclos médios cabem em 30 dias
pp = periodo_base / ciclo_medio
```

```python
# =============================================================
# VALOR DINÂMICO: LQ (Leads Qualificados Ativos)
# =============================================================
# Este é o ÚNICO valor que você altera no dia a dia.
# Representa quantos leads estão ativos agora no seu funil.
 

lq = 3
 
 
# =============================================================
# CALCULAR A RECEITA PREVISÍVEL
# =============================================================
 
# Receita estimada para UM ciclo médio de vendas
receita_ciclo = lq * e * tm
 
# Receita estimada para os próximos 30 dias
receita_30_dias = lq * e * tm * pp

print(f'R ciclo: R$ {math.floor(receita_ciclo)}, Receita de 30 dias: R$ {math.floor(receita_30_dias)}')
```
